```python
from __future__ import print_function, division
import pylab as pl
import pandas as pd
import numpy as np
import scipy as sp
import scipy.stats as stats
import scipy.stats
%pylab inline

import os
import json
```

```python
np.random.seed(14)
mymean = 500 
mystdev = 9
nSamples = 100
```

```python
#create a poisson distribution with some mean and standard deviation
x = sp.stats.poisson.rvs(mymean, size = nSamples)
```

```python
# Plot the origional distribution

pl.hist(x, color="grey")

pl.xlabel("Means")
pl.ylabel("Frequency")
pl.title("Poisson Distribution")
```

```python
mydis = sp.stats.poisson.rvs(mymean, size = nSamples)
myDmean = mydis.mean()
```

```python
nit = 1000
myDmeans = np.zeros(nit)
for i in range(nit):
    mydis = sp.stats.poisson.rvs(mymean, size = nSamples)
    myDmeans[i] = mydis.mean()
```

```python

pl.hist(mydis, color="grey")
pl.xlabel("Means")
pl.ylabel("Frequency")
pl.title("One Distribution")
```

```python
zs = (mymean - myDmeans) * np.sqrt(nSamples) / mystdev
zs.sort()


plt.hist(zs, color="grey")
```

```python
pl.hist(zs,normed=True, color="grey")

fit = stats.norm.pdf(zs, np.mean(zs), np.std(zs))
 
pl.plot(zs,fit, color="maroon")

pl.show
```

```python
# AD test
print ("poisson vs normal", scipy.stats.anderson(zs, dist='norm'))
```

## Null Hypothesis

### The sampled poisson distribution is drawn from a population that follows a gaussian distribution.


## Results of AD test:

### AD statistic = 0.338 which is lower then the alpha level(.05) statistic of 0.759 so you do not reject the null hypothesis
