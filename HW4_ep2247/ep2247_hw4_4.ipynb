```python
df = pd.read_csv('201501-citibike-tripdata.zip')
df.head()
```

```python

# df is the dataframe where the content of the csv file is stored
df['date'] = pd.to_datetime(df['starttime'])
# note that with dataframes I can refer to variables as dictionary keys, 
# i.e. df['starttime'] or as attributes: df.starttime. 
df.head()
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df.columns
```

```python

#df is the dataframe where the content of the csv file is stored
df['ageM'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]
```

```python
df.head()
```

# H0:
## there is no statistical difference between nighttime trips and daytime trips

### alpha = 0.05

```python
def day_or_night(hour):
    if hour >= 7 and hour <20:
        return 1
    else:
        return 0
    
df['DayorNight'] = df['hour'].apply(day_or_night)

df.head()
```

```python
df.describe()
```

##  h0:

### The age of bikers originating from Manhattan are younger then the age of bikers from Brooklyn

```python

# dropping NaN values
df['ageM'].dropna(inplace= True)
df['ageF'].dropna(inplace= True)
df['DayorNight'].dropna(inplace=True)
```

```python
df.describe()
```

### KS tests

```python
ks = scipy.stats.ks_2samp(df.tripduration[df.DayorNight == 0], df.tripduration[df.DayorNight == 1])
print(ks)
```

## ks test results

KS result statistic = 0.05 -> look at c(alpha) -> this number is 1.36 and our results should be greater than 1.36 to reject the null hypothesis.  The pvalue that i got running the test was 1.513e-117 which is close to 0.  This means that we can not reject the null hypothesis

# Pearson's test for correlation

```python
pc = scipy.stats.pearsonr(df.tripduration[df.DayorNight])
print(pc)
```
